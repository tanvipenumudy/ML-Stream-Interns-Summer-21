```python
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
#Reading the image
image = cv.imread('/content/123.jpg')
#Extracting the image height and width
h, w = image.shape[:2]
print("Height : " + str(h),  "Width : " +  str(w))
```

```python

# Extracting RGB values. 
# Here we have randomly chosen a pixel
# by passing in 250, 150 for height and width.
(B, G, R) = image[200, 150]
  
# Displaying the pixel values
print("R = {}, G = {}, B = {}".format(R, G, B))
  
```

```python

# We will calculate the region of interest 
# by slicing the pixels of the image
roi = image[250 : 600, 300 : 600]
cv2_imshow(roi)
```

```python

# resize() function takes 2 parameters, 
# the image and the dimensions
resize = cv.resize(image, (300, 300))
cv2_imshow(resize)
```

```python

# Calculating the center of the image
center = (w // 2, h // 2)
  
# Generating a rotation matrix
matrix = cv.getRotationMatrix2D(center, -45, 1.0) 
  
# Performing the affine transformation
rotated = cv.warpAffine(image, matrix, (h, w))
cv2_imshow(rotated)
```
