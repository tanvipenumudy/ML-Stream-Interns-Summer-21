```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
sns.relplot('Petal.Width','Petal.Length', hue ='Species', data = data)
```

```python
sns.relplot('Petal.Width','Petal.Length', data = data, kind = 'line', hue ='Species')
```

```python
sns.relplot('Sepal.Width','Sepal.Length', data = data, kind = 'line', col = 'Species' , hue = 'Species')
```

```python
sns.relplot('Sepal.Width','Sepal.Length', data = data, col = 'Species', kind = 'scatter', hue ='Petal.Width')
```

```python
sns.catplot(x ='Sepal.Width', y ='Sepal.Length', data = data, col = 'Species', kind ='box', hue ='Species')
```

```python
sns.lmplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, hue = 'Species', col = 'Species')
```

```python
sns.catplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, hue = 'Species', col = 'Species', kind = 'violin')
```

```python
sns.catplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, hue = 'Species', col = 'Species', kind = 'violin', inner = 'stick')
```

```python
sns.jointplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, hue = 'Species', kind = 'scatter')
```

```python
sns.jointplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, kind = 'hex')
```

```python
sns.jointplot(x = 'Sepal.Length', y = 'Sepal.Width', data = data, kind = 'kde')
```

```python
sns.pairplot(data)
```
