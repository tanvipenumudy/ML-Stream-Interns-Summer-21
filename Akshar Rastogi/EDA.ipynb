```python
def range_def(df ,column):
  rangereturn = []
  if (df [column]> 0) & (df [column] < 3).any():
    mark ='0-3'
    rangereturn.append(mark)
  elif (df [column] > 3) & (df [column] <5).any():
    mark ='3-5'
    rangereturn.append(mark)   
  elif (df [column] > 5) & (df [column] <7).any():
    mark ='5-7'
    rangereturn.append(mark)
  elif (df [column] > 7) &  (df [column] < 9).any():
    mark ='7-9'
    rangereturn.append(mark)          
  elif (df [column] > 9) & (df [column] <11).any():
    mark ='9-11'
    rangereturn.append(mark)     
  elif (df [column] > 11) & (df [column] < 14).any():
    mark ='11-14'
    rangereturn.append(mark)     
```
