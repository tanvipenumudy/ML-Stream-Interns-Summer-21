```python
img=cv2.imread('/content/drive/MyDrive/Internity - Summer Internship/internity.png')
imgG=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.waitKey(0)
plt.figure(figsize=(10,10))
plt.imshow(imgG)
plt.axis('off');
```

```python
imgBlur=cv2.GaussianBlur(img,(59,59),0)
plt.figure(figsize=(5,5))
plt.imshow(imgBlur)
plt.axis('off');
```

```python
imgCanny=cv2.Canny(img,100,100)
plt.figure(figsize=(5,5))
plt.imshow(imgCanny)
plt.axis('off');
```

```python
ones=np.ones((5,5),np.uint8)
imgDilate=cv2.dilate(imgCanny,ones,iterations=3)
plt.figure(figsize=(5,5))
plt.imshow(imgDilate)
plt.axis('off');
```

```python
imgerode=cv2.erode(imgDilate,ones,iterations=3)
plt.figure(figsize=(5,5))
plt.imshow(imgerode)
plt.axis('off');
```
