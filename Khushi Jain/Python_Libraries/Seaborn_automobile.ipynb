```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Automobile_data.csv')
data.head()
```

# Line Plot

```python
col1 = np.linspace(0, 10, 1000)
col2 = np.sin(col1)
col3 = np.cos(col1)
df = pd.DataFrame({"C1" : col1 , "C2" :col2, "C3": col3})
df.head(2)
```

```python
plt.style.use('seaborn-darkgrid')

sns.lineplot(x=df.C1,y=df.C2,data=df, label='sin(x)')
sns.lineplot(x=df.C1,y=df.C2+1,data=df, label='sin(x)+1')

plt.legend()
plt.show()
```

```python
# Change figure size
plt.figure(figsize=(5,3))

sns.lineplot(x=df.C1,y=df.C3,data=df)

sns.lineplot(x=df.C1,y=df.C3,data=df, label = 'cos')

plt.legend()
plt.show()
```

#### Create a palette

```python
# Create a palette from a single color

# Using RGB

color = (0.56, 0.87, 0.34)
sns.palplot(sns.light_palette(color))
sns.palplot(sns.dark_palette(color))

# And from a hex code:

sns.palplot(sns.color_palette("light:#124683"))
sns.palplot(sns.color_palette("dark:#124683"))
```

```python
# create a palette from named colors
sns.palplot(sns.blend_palette(["green", "red"], n_colors=4, input="xkcd"))
sns.palplot(sns.blend_palette(["red", "green"], n_colors=5, input="xkcd"))
```
