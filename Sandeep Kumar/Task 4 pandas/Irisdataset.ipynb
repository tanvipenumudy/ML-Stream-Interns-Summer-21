```python
#Dropping by position and not by label by:
df.drop(df.index[0],inplace=True)
#Remove multiple rows based on position by":
df.drop(df.index[[1,2]],inplace=True)
df.head()
```

```python
# Just like the select command in SQL and putting conditons alongwith
c1=df[df.flower_type=='Iris-setosa']
```
