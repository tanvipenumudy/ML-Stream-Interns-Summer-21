```python
le=LabelEncoder()
df=df.apply(le.fit_transform)
df.head()
```

```python
X=df.values[:,1:]
Y=df.values[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

# Building our classifier

```python
def prior_prob(y_train,label):
    total=y_train.shape[0]
    class_examples=np.sum(y_train==label)
    return (class_examples)/float(total)
```

```python
def cond_prob(x_train,y_train,col_index,col_val,label):
    x_filtered=x_train[y_train==label]
    numerator=np.sum(x_filtered[:,col_index]==col_val)
    denominator=np.sum(y_train==label)

    return numerator/float(denominator)
```

```python
def predict(x_train,y_train,xtest):
    """
    xtest is a single training point in x_test
    """
    classes=np.unique(y_train)
    num_of_features=x_train.shape[1]
    post_probs=[]
    for label in classes:
        likelihood=1.0
        for f in range(num_of_features):
            cond=cond_prob(x_train,y_train,f,xtest[f],label)
            likelihood*=cond
        prior=prior_prob(y_train,label)
        post=likelihood*prior
        post_probs.append(post)
    pred=np.argmax(post_probs)
    return pred
```

```python
y_pred=[predict(x_train,y_train,x_test[i]) for i in range(x_test.shape[0])]
len(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
