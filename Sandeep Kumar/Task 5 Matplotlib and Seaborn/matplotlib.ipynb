# Matplotlib Pyplot

**matplotlib.pyplot** is a collection of command style functions that are used to : 

Create a figure, Create a plotting area in a figure, Plot some lines in a plotting area, Decorate the plot with labels, etc. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([2,4, 6, 4])
plt.ylabel("Numbers")
plt.xlabel('Indices')
plt.title('MyPlot')
plt.show()
```

If we provide a single list or array ti the plot() then matplotlib assumes it is a sequence of y values and x axis values are assumed to be 0,1,2,3.

**Plot x versus y**

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.ylabel('squares')
plt.xlabel('numbers')
plt.grid() # Shows the grid here

plt.show()
```

The optional third aguemnt in the plot() function is the ***format string*** that indicates the color and line type of the plot
The bunch of format string can be referred from here: 
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
plt.plot([1,2,3,4],[1,4,9,16],'rs')
plt.grid()
plt.show()
```

```python
t=np.arange(0,5,0.2)

plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.grid() # To show th grid
plt.legend() #To show the legend of each line in the side of the graph
plt.show() # To plot the graph
```

```python
x=np.linspace(0,50,1000)
plt.plot(x,x**2,'--',linewidth=3.0)
plt.show()
```

```python
x1=np.arange(0,20,0.05)
# y1=[i*i for i in x1]
x2=np.arange(-20,0,0.05)
# y2=[i*i for i in x2]
lines=plt.plot(x1,x1*x1,x2,x2*x2,label="^2")
plt.setp(lines[0],color='r',linewidth='2.0')
plt.setp(lines[1],color='b',linewidth='3.0')
plt.legend()
plt.grid()
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)
# The Subplot command specifies numrows,nunmcols,fignum where fignum ranges form 1 to numrows*numcols

plt.subplot(211)
plt.grid()
plt.plot(t1,f(t1),'b')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi* t2),'r--')
plt.show()
```

```python
x=np.arange(1,50,5)

# plt.figure(1)
fig,axis=plt.subplots(3)

# plt.subplot(311)
axis[0].grid()
axis[0].plot(x,2*x,'--')

# plt.subplot(312)
axis[1].grid()
axis[1].plot(x,x*2,'r',label='x*2')
axis[1].plot(x,x*4,'g',label='x*4')
axis[1].legend()

# plt.subplot(313)
axis[2].grid()
lines=axis[2].plot(x,x,x,x*3)
plt.setp(lines[0],color='r',linewidth=1.5,label="x")
plt.setp(lines[1],color='b',linewidth=2,label="x*2")
axis[2].legend()

plt.show()

```
