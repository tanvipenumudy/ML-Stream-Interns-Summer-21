# Preprocessing

# Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading the dataset 

```python
df = pd.read_csv("train.csv")
```

```python
df.head() 
```

```python
df.info()
```

# Dropping columns which may not contribute much to our machine learning model

```python
cols = ['Name','Ticket','Cabin']
df = df.drop(['Name', 'Ticket', 'Cabin'],axis=1)
```

```python
df.info()
```

# Dropping rows having missing values

```python
df = df.dropna()
df.info()
```

# Convert Pclass, Sex, Embarked to dummy variables and drop them after conversion

```python
dummies = []
cols = ['Pclass','Sex','Embarked']
for col in cols:
    dummies.append(pd.get_dummies(df[col]))
```

```python
titanic_dummies = pd.concat(dummies, axis=1)
titanic_dummies.head()
```

```python
df = pd.concat((df,titanic_dummies),axis=1)
df.head()
```

```python
df = df.drop(['Pclass', 'Sex', 'Embarked'], axis=1)
df.head()
```

# Replacing the missing values by median using interpolate function.

```python
df['Age'] = df['Age'].interpolate()
```

# Splitting the dependent and the independent variables.

```python
X = df.values
y = df['Survived'].values
```

```python
X = np.delete(X, 1, axis=1)
```

# Splitting the data in test and training set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, random_state = 0)
```
